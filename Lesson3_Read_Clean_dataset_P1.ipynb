# **2. Series: Mảng một chiều**

**1. Pandas Series là gì?**

Series là cấu trúc dữ liệu cơ bản của thư viện Pandas. Bạn có thể hình dung nó như một mảng một chiều (1D array), tương tự như một cột duy nhất trong bảng tính Excel hoặc một danh sách (List) có gắn nhãn.

**2. Đặc điểm chính**

Dữ liệu đồng nhất: Một Series thường chứa các phần tử có cùng kiểu dữ liệu (integer, float, string...).

Trục chỉ số (Index): Mỗi phần tử trong Series được gắn với một nhãn (mặc định là số nguyên từ $0, 1, 2, ...$), giúp truy xuất dữ liệu cực nhanh.

```python
import pandas as pd

# Tạo Series từ list
tuoi = pd.Series([25, 30, 35, 40, 45])
print(tuoi)
print(type(tuoi))
```

```python
# Series với index tự định nghĩa
ten = pd.Series(
    ['An', 'Binh', 'Chi', 'Dung', 'Em'],
    index=['SV001', 'SV002', 'SV003', 'SV004', 'SV005']
)
print(ten)

# Truy cập
print(ten['SV001'])
print(ten[0])
```

# **3. DataFrame: Bảng dữ liệu hai chiều**

**1. Pandas DataFrame là gì?**

DataFrame là cấu trúc dữ liệu quan trọng nhất của Pandas. Bạn có thể hình dung nó là một bảng dữ liệu hai chiều (2D), có hàng và cột, hoàn toàn tương tự như một bảng tính Excel hoặc SQL table.

**2. Đặc điểm chính**

Đa chiều: Bao gồm nhiều cột dữ liệu gộp lại.

Linh hoạt: Mỗi cột trong một DataFrame có thể chứa một kiểu dữ liệu khác nhau (ví dụ: cột tên là chuỗi, cột điểm là số thực).

Cấu trúc: Gồm có Index (chỉ số hàng) và Columns (tên cột).

```python
import pandas as pd

# Tạo dữ liệu dưới dạng Dictionary
# Mỗi Key là tên cột, mỗi Value (List) là dữ liệu của cột đó
data = {
    'Ten': ['An', 'Binh', 'Chi', 'Dung', 'Em'],
    'Tuoi': [25, 30, 35, 40, 45],
    'Thanh_pho': ['HN', 'HCM', 'DN', 'HN', 'HCM'],
    'Luong': [10, 15, 12, 18, 20]
}

# Khởi tạo DataFrame từ dictionary trên
df = pd.DataFrame(data)

# In DataFrame ra màn hình
print("Dữ liệu bảng DataFrame:")
print(df)

# Gợi ý: Trong Colab, bạn có thể dùng lệnh display(df) để nhìn bảng đẹp hơn
# display(df)
```

5.4.1. Chọn cột: Dot or Bracket

```python
# 1. Sử dụng Dấu ngoặc vuông (Bracket notation) - CÁCH NÊN DÙNG
# Chọn 1 cột (Kết quả trả về là một Series)
ten_kh = df_sample['Ten']

# Chọn NHIỀU cột cùng lúc (Kết quả trả về là một DataFrame)
# Lưu ý: Sử dụng ngoặc vuông kép [[...]]
df_con = df_sample[['Ten', 'Thanh_pho', 'Thu_nhap']]

print("Chọn bằng ngoặc vuông (nhiều cột):")
display(df_con.head(3))
```

5.4.2. Chọn hàng: loc và iloc

```python
# --- 1. Chọn Hàng (Rows) ---

# Chọn hàng đầu tiên bằng vị trí (iloc)
row_first = df_sample.iloc[0]
print(row_first.tail(3))

# Chọn hàng có nhãn là 0 (loc) - Mặc định index là số nếu không đặt lại
row_label_0 = df_sample.loc[0]
print(row_first.sample(3))
```

```python
# --- 2. Chọn Hàng và Cột cụ thể ---

# Cấu trúc: df.loc[hàng, cột] hoặc df.iloc[hàng, cột]

# Lấy Tên và Thu nhập của khách hàng đầu tiên (dùng loc)
info_loc = df_sample.loc[0, ['Ten', 'Thu_nhap']]
print(info_loc.head())
# Lấy giá trị ở hàng 0, cột 1 (dùng iloc)
info_iloc = df_sample.iloc[0, 1]
print(info_iloc)
```

```python
# --- 3. Cắt lớp (Slicing) ---

# Lấy từ hàng 5 đến hàng 10 (iloc không lấy điểm cuối: 5 đến 9)
slice_iloc = df_sample.iloc[5:10,:]
print(slice_iloc)

# Lấy từ hàng 5 đến hàng 10 (loc lấy CẢ điểm cuối: 5 đến 10)
slice_loc = df_sample.loc[5:10]
print(slice_loc)
# Cắt lớp cả hàng và cột
# Lấy 5 hàng đầu và 3 cột đầu tiên
sub_df = df_sample.iloc[0:5, 0:3]
print(sub_df)
```

5.4.3. Lọc theo điều kiện

```python
# 1. Lọc khách hàng dưới 30 tuổi
df_tre = df_sample[df_sample['Tuoi'] < 30]
print(f"Số người dưới 30 tuổi: {len(df_tre)}")

# 2. Lọc khách hàng ở 'HN' VÀ có thu nhập > 30,000,000
# Lưu ý: Bắt buộc dùng ngoặc đơn () cho mỗi điều kiện
df_hn_giau = df_sample[(df_sample['Thanh_pho'] == 'HN') & (df_sample['Thu_nhap'] > 30000000)]
print(f"Số người ở HN có thu nhập cao: {len(df_hn_giau)}")

# 3. Lọc khách hàng ở các thành phố miền Bắc (HN hoặc HP) dùng .isin()
thanhpho_mien_bac = ['HN', 'HP']
df_mien_bac = df_sample[df_sample['Thanh_pho'].isin(thanhpho_mien_bac)]
print(f"Số người ở thành phố miền Bắc: {len(df_mien_bac)}")

# 4. Phủ định: Lọc khách hàng KHÔNG ở 'HCM'
df_khong_hcm = df_sample[df_sample['Thanh_pho'] != 'HCM']
# Hoặc dùng dấu ngã (~): df_sample[~df_sample['Thanh_pho'].isin(['HCM'])]

# Xem thử kết quả của nhóm HN giàu
display(df_hn_giau.head())
```

**THỰC HÀNH 3: TRUY XUẤT DỮ LIỆU**

Thực hiện các câu hỏi truy vấn để hiểu sâu về dữ liệu đã tạo:

Thống kê số lượng: Đếm xem có bao nhiêu người ở một thành phố cụ thể (ví dụ: Hà Nội).

Tính toán trung bình: Tính thu nhập trung bình của nhóm bạn.

Tìm giá trị cực trị: Xác định ai là người trẻ nhất và ai là người già nhất trong danh sách.

Sắp xếp dữ liệu: Tìm ra Top những người có mức lương cao nhất.

Tính tỷ lệ phần trăm: Tính xem có bao nhiêu % người trong nhóm có mức lương trên một ngưỡng nhất định (ví dụ: > 30 triệu).

```python
# 1. Có bao nhiêu người ở HCM?
# Sử dụng df_sample thay cho df
so_nguoi_hcm = len(df_sample[df_sample['Thanh_pho'] == 'HCM'])
print(f"1. Số người ở HCM: {so_nguoi_hcm} người")

# 2. Thu nhập trung bình của người HN?
# Lọc theo 'Thanh_pho' == 'HN' sau đó tính mean của cột 'Thu_nhap'
thu_nhap_tb_hn = df_sample[df_sample['Thanh_pho'] == 'HN']['Thu_nhap'].mean()
print(f"2. Thu nhập TB tại HN: {thu_nhap_tb_hn:,.0f} VND")

# 3. Người trẻ nhất và già nhất?
# Sử dụng hàm min() và max() trên cột 'Tuoi'
print(f"3. Trẻ nhất: {df_sample['Tuoi'].min()} tuổi")
print(f"   Già nhất: {df_sample['Tuoi'].max()} tuổi")

# 4. Top 5 người có thu nhập cao nhất?
# Sắp xếp giảm dần theo 'Thu_nhap' và lấy 5 dòng đầu
print("\n4. Top 5 người có thu nhập cao nhất:")
top_5_thu_nhap = df_sample.sort_values('Thu_nhap', ascending=False).head(5)
display(top_5_thu_nhap)

# 5. Bao nhiêu % người có thu nhập > 30 triệu?
# (Số người thỏa điều kiện / Tổng số người) * 100
ti_le_tren_30tr = (len(df_sample[df_sample['Thu_nhap'] > 30000000]) / len(df_sample)) * 100
print(f"\n5. Tỉ lệ người thu nhập > 30tr: {ti_le_tren_30tr:.1f}%")
```

# **6. XỬ LÝ DỮ LIỆU**

**6.1. Phát hiện dữ liệu thiếu**

**THỰC HÀNH 5: lÀM SẠCH DỮ LIỆU**

Mục tiêu: Thực hiện quy trình làm sạch dữ liệu (Data Cleaning) trên bộ dữ liệu Titanic để chuẩn bị cho việc phân tích hoặc huấn luyện mô hình máy tính.

Kiểm tra dữ liệu thiếu (Missing values): Xác định các cột nào đang bị trống dữ liệu và số lượng là bao nhiêu.

Xử lý cột thiếu quá nhiều: Nếu một cột có tỉ lệ dữ liệu thiếu trên 50%, ta nên loại bỏ hoàn toàn cột đó (ví dụ cột Cabin).

Điền dữ liệu thiếu cho cột 'Age' (Tuổi): Không dùng giá trị trung bình chung mà phải điền giá trị Trung vị (Median) theo từng Hạng vé (Pclass) để đảm bảo tính logic (khách hạng sang thường có độ tuổi khác khách hạng phổ thông).

Điền dữ liệu thiếu cho cột 'Embarked' (Cổng lên tàu): Đây là dữ liệu dạng chữ, ta điền bằng giá trị xuất hiện nhiều nhất (Mode).

Xử lý các dòng còn sót lại: Sau khi đã điền các cột chính, nếu vẫn còn dòng nào chứa ô trống thì xóa bỏ dòng đó.

Xác nhận kết quả: Kiểm tra lại lần cuối để đảm bảo số lượng giá trị thiếu bằng 0.

```python
import pandas as pd

# ==========================================
# 0. Tải bộ dữ liệu Titanic từ URL
# ==========================================
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df_titanic = pd.read_csv(url)
print("Dữ liệu gốc:")
display(df_titanic.head())
```

```python
# ==========================================
# 1. Kiểm tra missing values ban đầu
# ==========================================
print("--- Bước 1: Thống kê dữ liệu thiếu ban đầu ---")
print(df_titanic.isnull().sum())
print("-" * 40)
```

```python
# ==========================================
# 2. Xóa cột có > 50% missing (Thường là cột Cabin)
# ==========================================
# Tính ngưỡng 50% số dòng
threshold = len(df_titanic) * 0.5
# Lọc các cột có số lượng null vượt ngưỡng
cols_to_drop = df_titanic.columns[df_titanic.isnull().sum() > threshold]
df_titanic = df_titanic.drop(columns=cols_to_drop)
print(f"Bước 2: Đã xóa các cột thiếu > 50%: {list(cols_to_drop)}")
```

```python
# ==========================================
# 3. Điền Age bằng median theo Pclass
# ==========================================
# Dùng transform để tính median cho từng nhóm Pclass rồi điền vào Age
df_titanic['Age'] = df_titanic['Age'].fillna(
    df_titanic.groupby('Pclass')['Age'].transform('median')
)
print("Bước 3: Đã điền Age bằng median theo từng Pclass.")
```

```python
# ==========================================
# 4. Điền Embarked bằng mode (Giá trị phổ biến nhất)
# ==========================================
# mode() trả về một Series, ta lấy giá trị đầu tiên [0]
most_common_embarked = df_titanic['Embarked'].mode()[0]
df_titanic['Embarked'] = df_titanic['Embarked'].fillna(most_common_embarked)
print(f"Bước 4: Đã điền Embarked bằng giá trị phổ biến nhất: {most_common_embarked}")
```

```python
# ==========================================
# 5. Xóa các rows còn lại có missing
# ==========================================
df_titanic = df_titanic.dropna()
print("Bước 5: Đã xóa các dòng còn sót lại có giá trị thiếu.")
print(df_titanic)
```

```python
# ==========================================
# 6. Kiểm tra lại - phải đạt 0 missing!
# ==========================================
print("\n--- Bước 6: Kiểm tra lại tổng số missing values ---")
final_missing_count = df_titanic.isnull().sum()
print(final_missing_count)

if final_missing_count.sum() == 0:
    print("\n✅ KẾT QUẢ: Dữ liệu đã sạch hoàn toàn (0 missing)!")
else:
    print("\n⚠️ CẢNH BÁO: Vẫn còn dữ liệu thiếu.")

# Hiển thị 5 dòng đầu của dữ liệu sau khi xử lý
print("\nDữ liệu sau khi làm sạch:")
print(df_titanic.head())
```

Kiểm chứng dữ liệu (Validate): Đảm bảo tuổi từ 18-100 và lương > 0.

So sánh trước và sau (Before/After): Sử dụng hàm báo cáo chất lượng để đánh giá hiệu quả sau khi làm sạch.

```python
import numpy as np
import pandas as pd

# --- BƯỚC 0: TẠO MESSY DATASET (GIẢ LẬP THỰC TẾ) ---
np.random.seed(42)
n = 100

df_messy = pd.DataFrame({
    'ID': range(1, n+1),
    'Name': [f'Customer_{i}' if i % 10 != 0 else np.nan for i in range(1, n+1)],
    'Age': np.random.choice([np.nan, *range(18, 70)], n),
    'City': np.random.choice(['HN', 'HCM', 'hn', 'hcm', np.nan], n),
    'Salary': np.random.choice([np.nan, *range(5000000, 50000000, 1000000)], n)
})

# Thêm dữ liệu trùng lặp (duplicates)
df_messy = pd.concat([df_messy, df_messy.iloc[10:15]], ignore_index=True)
```

```python
# --- BƯỚC 1: VIẾT HÀM BÁO CÁO CHẤT LƯỢNG ---
def data_quality_report(df):
    report = {
        "Total Rows": df.shape[0],
        "Total Columns": df.shape[1],
        "Missing Values (%)": df.isnull().mean() * 100,
        "Duplicates": df.duplicated().sum(),
        "Unique Values": df.nunique()
    }
    return report

print("--- Báo cáo chất lượng dữ liệu TRƯỚC khi làm sạch ---")
report_before = data_quality_report(df_messy)
for key, value in report_before.items():
    print(f"{key}:\n{value}\n")

# --- BƯỚC 2: CLEAN DATA ---
df_clean = df_messy.copy()

# 2.1 Xóa duplicates
df_clean = df_clean.drop_duplicates()

# 2.2 Chuẩn hóa City
df_clean['City'] = df_clean['City'].str.lower()
city_mapping = {'hn': 'Hà Nội', 'hcm': 'TP.HCM'}
df_clean['City'] = df_clean['City'].replace(city_mapping)

# 2.3 Xử lý missing values
# - Xóa các dòng không có Tên hoặc Thành phố (vì khó giả lập)
df_clean = df_clean.dropna(subset=['Name', 'City'])
# - Điền giá trị trung bình cho Age và Salary
df_clean['Age'] = df_clean['Age'].fillna(df_clean['Age'].median())
df_clean['Salary'] = df_clean['Salary'].fillna(df_clean['Salary'].mean())

# 2.4 Validate dữ liệu
df_clean = df_clean.loc[(df_clean['Age'] >= 18) & (df_clean['Age'] <= 100)]
df_clean = df_clean.loc[df_clean['Salary'] > 0]

# --- BƯỚC 3: SO SÁNH BEFORE/AFTER ---
print("--- Báo cáo chất lượng dữ liệu SAU khi làm sạch ---")
report_after = data_quality_report(df_clean)
for key, value in report_after.items():
    print(f"{key}:\n{value}\n")

print(f"Số lượng bản ghi bị loại bỏ: {len(df_messy) - len(df_clean)}")
```
